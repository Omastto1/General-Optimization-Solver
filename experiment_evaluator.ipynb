```python
optimum = 0
feasible = []
worse_than_optimum = []
worse_than_upper = []
worse_than_lower = []
not_known = 0
for instance in instances:
    if instance["run_history"][-1]["solve_status"] == "Optimal":
        optimum += 1
    else:
        if instance["reference_solution"]["optimum"] is not None:
            ratio = round((instance["run_history"][-1]["solution_value"] / instance["reference_solution"]["optimum"] - 1) * 100, 1)
            print(f"Solution is {ratio} % worse than the optimum.")
            worse_than_optimum.append(ratio)
        elif instance["reference_solution"]["bounds"] is not None:
            if instance["run_history"][-1]["solution_value"] >= instance["reference_solution"]["bounds"]["upper"]:
                ratio = round((instance["run_history"][-1]["solution_value"] / instance["reference_solution"]["bounds"]["upper"] - 1) * 100, 1)
                print(f"Solution is {ratio} % worse than the upper bound.")
                worse_than_upper.append(ratio)
            else:
                ratio = round((instance["run_history"][-1]["solution_value"] / instance["reference_solution"]["bounds"]["lower"] - 1) * 100, 1)
                print(f"Solution is {ratio} % worse than the lower bound.")
                worse_than_lower.append(ratio)
        else:
            print("There in no known reference solution in current data")
            not_known += 1
    
    # solve_times.append(instance["run_history"][-1]["solve_time"])
    # reference_solve_times.append(instance["reference_solution"]["cpu_time"])

print("Number of optimally run instances: ", optimum)
print("Number of instances that are worse than the optimum: ", len(worse_than_optimum))
if len(worse_than_optimum) > 0:
    print(f"average gap to the optimum: {round(sum(worse_than_optimum) / len(worse_than_optimum), 1)} %")
print("Number of instances that are worse than the upper bound: ", len(worse_than_upper))
if len(worse_than_upper) > 0:
    print(f"average gap to the upper bound: {round(sum(worse_than_upper) / len(worse_than_upper), 1)} %")
print("Number of instances that are worse than the lower bound: ", len(worse_than_lower))
if len(worse_than_lower) > 0:
    print(f"average gap to the lower bound: {round(sum(worse_than_lower) / len(worse_than_lower), 1)} %")
```

```python

import matplotlib.pyplot as plt
```

```python
def visualize_progress(search_progress, instance):
    numbers = [entry[0] for entry in search_progress]
    times = [entry[2] for entry in search_progress]

    best = instance['reference_solution']['optimum']
    our_best_solution = search_progress[-1][0]
    if best and best == our_best_solution:
        best_known_solution = float(best)
        plt.axhline(y=best_known_solution, color='r', linestyle='--', label='Best Known Solution')
    
    if our_best_solution and best and best != our_best_solution:
        plt.axhline(y=our_best_solution, color='g', linestyle='--', label='Our Best Solution')

    # Plot the data
    plt.plot(times, numbers, 'bo-', label='Results')
    plt.ylabel('Make span')
    plt.xlabel('Time (seconds)')
    plt.title('CV25 Solution progress')
    plt.legend()

    # plt.ylim(min(times), max(times + [best_known_solution]))

    # Display the plot
    plt.show()
```
